```python
def _format_historical_data(data):
        """Format historical data to standardized format."""
        # Chuyển đổi và loại bỏ key ticker cho mỗi dict trong data
        for item in data:
            item.pop('ticker')
            item['time'] = item['time'].strftime('%Y-%m-%d')
        return {
            "data": data,
            "total_records": len(data),
            "status": "success"
        }
```

```python
from typing import Dict, Any, List, Optional
def _safe_get_numeric(self, data: Dict[str, Any], keys: List[str]) -> Optional[float]:
        """Safely get numeric value from data using multiple possible keys."""
        for key in keys:
            if key in data:
                value = data[key]
                if pd.notna(value):
                    try:
                        return float(value)
                    except (ValueError, TypeError):
                        continue
        return None
```

```python
from typing import Dict, Any
def _format_financial_ratios(ratios_data: Any, symbol: str) -> Dict[str, Any]:
        """Format financial ratios from vnstock's financial_ratio output."""
        try:
            transposed_ratios = ratios_data.set_index('range').T
            latest_ratios = transposed_ratios.iloc[:, 0].to_dict()
            formatted_data = {
                "pe": _safe_get_numeric(latest_ratios, ["priceToEarning", "PE"]),
                "pb": _safe_get_numeric(latest_ratios, ["priceToBook", "PB"]),
                "roe": _safe_get_numeric(latest_ratios, ["roe", "returnOnEquity", "ROE"]),
                "eps": _safe_get_numeric(latest_ratios, ["earningPerShare", "EPS"]),
                "market_cap": None,
                "revenue": _safe_get_numeric(latest_ratios, ["totalRevenue"]),
                "profit": _safe_get_numeric(latest_ratios, ["netIncome"]),
                "debt_ratio": _safe_get_numeric(latest_ratios, ["debtOnEquity", "equityOnLiability"]),
                "current_ratio": _safe_get_numeric(latest_ratios, ["currentRatio"]),
                "quick_ratio": _safe_get_numeric(latest_ratios, ["quickRatio"]),
                "gross_margin": _safe_get_numeric(latest_ratios, ["grossMargin"]),
                "net_margin": _safe_get_numeric(latest_ratios, ["postTaxOnToi", "netMargin"]),
                "roa": _safe_get_numeric(latest_ratios, ["roa", "returnOnAssets", "ROA"]),
                "symbol": symbol,
                "timestamp": datetime.now().isoformat(),
                "status": "success"
            }
            return formatted_data
        except Exception as e:
            print(f"Failed to format fundamental ratios for {symbol}: {str(e)}")
```

```python
def _format_stock_ls_analysis(df: pd.DataFrame) -> Dict[str, Any]:
    """Formats the stock_ls_analysis DataFrame to a dict with ticker as key and indicators as values."""
    # column_mapping = {
    #     'marcap': 'Vốn hóa (tỷ)',
    #     'price': 'Giá',
    #     'numberOfDays': 'Số phiên tăng/giảm liên tiếp',
    #     'priceToEarning': 'P/E',
    #     'peg': 'PEG',
    #     'priceToBook': 'P/B',
    #     'dividend': 'Cổ tức',
    #     'roe': 'ROE',
    #     'roa': 'ROA',
    #     'debtOnEquity': 'Nợ/Vốn CSH',
    #     'income5year': 'LNST 5 năm',
    #     'sale5year': 'Doanh thu 5 năm',
    #     'income1quarter': 'LNST quý gần nhất',
    #     'sale1quarter': 'Doanh thu quý gần nhất',
    #     'nextIncome': 'LNST năm tới',
    #     'nextSale': 'Doanh thu năm tới',
    #     'rsi': 'RSI',
    #     'rs': 'RS'
    # }
    
    # # Đổi tên index theo column_mapping, loại bỏ các index không liên quan nếu có
    # df = df.rename(index=column_mapping)
    
    # Lấy danh sách các chỉ số (index) từ 'Vốn hóa (tỷ)' trở đi
    indicators = [idx for idx in df.index]
    
    # Chuẩn bị dict kết quả
    result = {}
    
    # Lặp qua từng cột (mã chứng khoán)
    for ticker in df.columns:
        ticker_data = {}
        for indicator in indicators:
            value = df.loc[indicator, ticker]
            if pd.notna(value):
                if indicator == 'Vốn hóa (tỷ)':
                    ticker_data[indicator] = int((value / 1e3)) if value > 0 else 0
                elif isinstance(value, (int, float)):
                    ticker_data[indicator] = round(value, 2)
                else:
                    ticker_data[indicator] = value
        result[ticker] = ticker_data

    return result
```
